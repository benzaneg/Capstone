## Cambridge Scrapper

```python
def cambridge_histories_scraper_af(url_link, list_of_chapters):    
       
    # Open up the Selenium Chrome Webdriver
    browser = webdriver.Chrome(executable_path = "./chromedriver/macos/chromedriver")
    path="./chromedriver/macos/chromedriver"
        # Navigate through the log in process
    browser.get(url_link)
    time.sleep(1)
    browser.find_element_by_id("weblogin_netid").send_keys('idssam') # need to remove password
    time.sleep(1)
    browser.find_element_by_id("weblogin_password").send_keys('Ret@2017') # need to remove password
    time.sleep(1)
    browser.find_element_by_id("submit_button").click()

    # Grabbing an HTML version of the current page
    html_doc = browser.page_source
    time.sleep(1)
    soup = BeautifulSoup(html_doc, 'html.parser')
        # Define the Corpora
    corpora = []

    # Creating a list of all of the chapter URL links I need to scrape 
    for tag in soup.find_all(class_ = 'category'):
        chap_titles = []
        links = []
        for piece in tag.find_all('a'):
            links.append(piece['href'])
            chap_titles.append(piece.text)
    new_links = []
    for item in links:
        if "/core/books/new-cambridge-history-of-american-foreign-relations/" in item:
            almost = str(re.findall('relations\/([0-9a-z-]+\/[A-Z0-9]+)', item))
            new_links.append(almost)
            
    cleaned = []
    for dirty in new_links:
        dirty = re.sub("\['", '', dirty)
        dirty = re.sub("'\]", '', dirty)
        dirty = dirty + "/core-reader"
        cleaned.append(dirty)
      
            
    # Creating chap titles for column headers later on
    clean_chap = []
    for item in chap_titles:
        if re.findall('\\n\d+', item):
            item = re.sub('pp[\s\d-]+', '', item)
            item = re.sub('\d[\s-]+', '', item)
            item = re.sub('\n', '', item)
            item = item.replace(' ', "_")
            clean_chap.append(item)
    print(clean_chap)    
    
    # Need to Select the right chapter titles 
    final_chap_titles = []
    for chap in (range(len(clean_chap))):
        final_chap_titles.append(clean_chap[chap])
        print(clean_chap[chap])

    
    # Now scraping the choosen, cleaned, chapter links
    for clean in list_of_chapters:
            browser = webdriver.Chrome(executable_path = "./chromedriver/macos/chromedriver")
            path="./chromedriver/macos/chromedriver"
            url = 'https://www-cambridge-org.offcampus.lib.washington.edu/core/books/new-cambridge-history-of-american-foreign-relations/{}'.format(cleaned[clean])
            browser.get(url)
            time.sleep(1)
            browser.find_element_by_id("weblogin_netid").send_keys('idssam') # need to remove password
            time.sleep(1)
            browser.find_element_by_id("weblogin_password").send_keys('Ret@2017') # need to remove password
            time.sleep(1)
            browser.find_element_by_id("submit_button").click()
            html_doc = browser.page_source
            time.sleep(1)
            soup = BeautifulSoup(html_doc, 'html.parser')
            corpus = [tag.text for tag in soup.find_all(class_ = 'p')]
            time.sleep(2)
            browser.close()
            corpora.append(corpus)
    
    corpora_df = pd.DataFrame(corpora).T
    corpora_df.columns = final_chap_titles
    
    return corpora_df, corpora
```

```python
# Creates a list of all the individual paragraphs 
list_paragraphs = []
for corp in corpora_arr_v1:
    for section in corp:
        list_paragraphs.append(section)
for corp in corpora_arr_v2:
    for section in corp:
        list_paragraphs.append(section)
for corp in corpora_arr_v3:
    for section in corp:
        list_paragraphs.append(section)
for corp in corpora_arr_v4:
    for section in corp:
        list_paragraphs.append(section)    

# Grabs all paragrpahs with a date in it, if no date fills with empty bracket
dates = []
for para in list_paragraphs:
    doc = nlp(para)
    dated = []
    for ent in doc.ents:
        if ent.label_ == "DATE":
            dated.append(ent.text)
    dates.append(dated)

### Data Frame One made from list of paragraphs which is a list of total paragrapghs

# Creates a dataframe with the paragraphs in it which I then add the dates onto
df = pd.DataFrame(list_paragraphs, columns = ['para'])
df['dates'] = dates

# Remove the paragraphs which are citations
count = 0
count_num = []
for x in df['para']:
    count +=1
    if x in re.findall('(^\d.+)', x):
        count_num.append(count)
real_count_num = []
for x in count_num:
    real_count_num.append(x - 1)
    
df = df.drop(df.index[real_count_num])    

# Creating a lemma version of chapters
length_para = []
para_lemma = []
for chap in df['para']:
    sent = chap.lower()
    sent = nlp(sent)
    sent = [token for token in sent if token.is_stop == False]
    sent = [token.lemma_.lower().strip() for token in sent]
    sent = [token for token in sent if token not in punctuations]
    length_para.append(len(sent))
    para_lemma.append(' '.join(sent))    

df['para_lemma'] = para_lemma
df['len'] = length_para

#Create a datadframe which just houses paragraphs with dates in it
dated_rows = []
for num in range(len(df['dates'])):
    if len(df['dates'].iloc[num]) > 0:
          dated_rows.append(df.iloc[num])
date_text = pd.DataFrame(dated_rows).reset_index(drop = True)
date_text.head()

### Data Frame Two made from this exchange which is all of the paragraphs that have date labels

# Creates a row where a four digit year is delmited
year = []
for row in date_text['dates']:
    new_row = []
    for item in row:
        if len(re.findall('[\d]{4}', item)) > 0:
            new_row.append(int(''.join(re.findall('[\d]{4}', item))))
    year.append(new_row)
date_text['year'] = year
date_text.head()

print("Shape of date_text df: ", date_text.shape)

### Data Frame Three made from this exchange where these are the paragraphs with four digit dates

# Creates a dataframe where there is four digits in it
year_text = []
for num in range(len(date_text['year'])):
    if len(date_text['year'].iloc[num]) > 0:
          year_text.append(date_text.iloc[num])
year_text = pd.DataFrame(year_text).reset_index(drop = True)
year_text.head()


# How will I know if they talk about an era such as 1800s or mid 1850s, need to look into

# Gives just a list of numbers, 4109 of them, with no application specifcially, trying to condense...
# Create a datadframe which just houses paragraphs with dates in it
# dated_rows = []
# for num in range(len(df['dates'])):
#     if len(df['dates'].iloc[num]) > 0:
#           for item in df['dates'].iloc[num]:
#                 if len(re.findall('[\d]{4}', item)) > 0:
#                     row_dates = []
#                     row_dates.append(int(''.join(re.findall('[\d]{4}', item))))
#                     dated_rows.append(row_dates)
# dated_rows = pd.DataFrame(dated_rows).reset_index(drop = True)
# dated_rows.head()
```

## Model from Chapter DF

```python
def print_top_words(model, feature_names, n_top_words = 20):
    for topic_idx, topic in enumerate(model.components_):
        message = "Topic #%d: " % topic_idx
        message += ", ".join([feature_names[i]
                             for i in topic.argsort()[:-n_top_words - 1:-1]])
        print(message)
    print()
#CountVectorizer(max_features=1000,stop_words='english', max_df = .7, min_df = 8, ngram_range = (1, 3))
    
# Use tf (raw term count) features for LDA.
print("Extracting tf features for LDA...")
tf_vectorizer = CountVectorizer(ngram_range = (1, 3),
                                max_df = 0.7,
                               min_df = 8,
                               #max_features = 1000
                               )
t0 = time()
tf = tf_vectorizer.fit_transform(chapters_all['chapter_lemma'])
print("done in %0.3fs." % (time() - t0))

# print("Fitting LDA models with tf features, "
#       "n_samples=%d and n_features=%d..."
#       % (n_samples, n_features))
lda = LatentDirichletAllocation(n_components=4, learning_decay = .7, max_iter = 25)
t0 = time()
lda.fit(tf)
print("done in %0.3fs." % (time() - t0))

print("Number of words: ", len(tf_vectorizer.vocabulary_))
print('Perplexity: ', lda.perplexity(tf))
print("\nTopics in LDA model:")
tf_feature_names = tf_vectorizer.get_feature_names()

lda_sim = pd.DataFrame(cosine_similarity(lda.components_, lda.components_))
sns.heatmap(lda_sim)

pickle.dump(lda ,open('lda_chap4.txt', 'wb'))
lda_n_grams = pickle.load(open('lda_chap4.txt', 'rb'))
print_top_words(lda_n_grams, tf_feature_names)

pyLDAvis.sklearn.prepare(lda, tf, tf_vectorizer)
```

```python
bigram = gensim.models.Phrases(chapters_all['chapter_gensim_lemma'], min_count=5, threshold=100) # higher threshold fewer phrases.
trigram = gensim.models.Phrases(bigram[chapters_all['chapter_gensim_lemma']], threshold=100)

bigram_mod = gensim.models.phrases.Phraser(bigram)
trigram_mod = gensim.models.phrases.Phraser(trigram)

def make_bigrams(texts):
    return [bigram_mod[doc] for doc in texts]

chap_words_bigrams = make_bigrams(chapters_all['chapter_gensim_lemma']) 

# Create Dictionary
id2word = corpora.Dictionary(chap_words_bigrams)

# Create Corpus
texts = chap_words_bigrams

# Term Document Frequency
corpus = [id2word.doc2bow(text) for text in texts]
```
